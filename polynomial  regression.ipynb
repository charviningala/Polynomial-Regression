```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x=np.linspace(-5,5,50)
y=6*x**3+9*x**2+1
z=np.random.normal(1,70,50)
p=y+z
plt.scatter(x,y,marker='+',color='red',label='without noise')

plt.scatter(x,p,marker='*',color='blue',label='with noise')
plt.legend()
plt.title('X vs Y')
plt.grid(True)
plt.show()
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(x.reshape(-1, 1))
poly.fit(x.reshape(-1,1))
```

```python
poly.transform(x.reshape(-1,1))
```

```python
degree = 4  # Set your desired polynomial degree
poly = PolynomialFeatures(degree=degree)

poly_reg_model = LinearRegression()
poly_reg_model.fit(poly_features, y)
y_predicted = poly_reg_model.predict(poly_features)

poly_r2 = poly_reg_model.score(poly_features, y)
print(f"Polynomial Regression (Degree {degree}) R² Score: {poly_r2:.5f}")


y_test_pred = poly_reg_model.predict(poly_features)


```
